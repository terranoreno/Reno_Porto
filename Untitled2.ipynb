```python
# Deep Insight: Does Test Preparation Increase Chance of Passing All Subjects?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Load and clean data
df = pd.read_csv('StudentsPerformance.csv')
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Rename for clarity
df['test_preparation_course'] = df['test_preparation_course'].replace({
    'none': 'No Prep', 'completed': 'Completed'
})

# Create a new column: passed_all = 1 if score > 70 in all subjects
df['passed_all'] = ((df['math_score'] > 70) & 
                    (df['reading_score'] > 70) & 
                    (df['writing_score'] > 70)).astype(int)

# Group by test prep
prep_pass_rate = df.groupby('test_preparation_course')['passed_all'].mean().reset_index()
print(prep_pass_rate)

# Barplot: Passing Rate Comparison
sns.barplot(data=prep_pass_rate, x='test_preparation_course', y='passed_all', palette='Set2')
plt.title("Pass Rate in All Subjects by Test Preparation")
plt.ylabel("Proportion of Students Passed All (>70)")
plt.ylim(0, 1)
plt.show()

# Contingency Table
contingency = pd.crosstab(df['test_preparation_course'], df['passed_all'])

# Chi-square test
chi2, pval, dof, expected = chi2_contingency(contingency)
print(f"Chi-square: {chi2:.2f}, p-value: {pval:.4f}")

if pval < 0.05:
    print("✅ Statistically significant difference: Test prep improves chance of passing.")
else:
    print("⚠️ No significant evidence: Test prep may not make a difference.")
```

```python
# Deep Insight: Student Performance Risk Segmentation

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load & clean
df = pd.read_csv('StudentsPerformance.csv')
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Hitung total skor
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']

# Buat kategori risiko (threshold bisa disesuaikan)
def categorize_risk(score):
    if score >= 240:
        return 'Low Risk'
    elif score >= 180:
        return 'Medium Risk'
    else:
        return 'High Risk'

df['risk_level'] = df['total_score'].apply(categorize_risk)

# Lihat distribusi risiko
risk_counts = df['risk_level'].value_counts(normalize=True)
print("Distribusi Risiko Akademik:")
print(risk_counts)

# Visualisasi proporsi risiko
sns.countplot(x='risk_level', data=df, order=['High Risk', 'Medium Risk', 'Low Risk'], palette='coolwarm')
plt.title("Distribusi Risiko Akademik Siswa")
plt.ylabel("Jumlah Siswa")
plt.xlabel("Level Risiko")
plt.show()

# Analisis demografi berdasarkan level risiko
cross_gender = pd.crosstab(df['risk_level'], df['gender'], normalize='index') * 100
cross_lunch = pd.crosstab(df['risk_level'], df['lunch'], normalize='index') * 100
cross_prep = pd.crosstab(df['risk_level'], df['test_preparation_course'], normalize='index') * 100
cross_parent_edu = pd.crosstab(df['risk_level'], df['parental_level_of_education'], normalize='index') * 100

# Tampilkan satu contoh
print("Distribusi Gender Berdasarkan Risiko:")
print(cross_gender.round(1))

# Visualisasi gender vs risk
cross_gender.plot(kind='bar', stacked=True, colormap='Set2')
plt.title("Gender Distribution by Academic Risk Level")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.show()
```
